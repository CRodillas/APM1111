<h2>1. (5 points). Table 1 shows a frequency distribution of grades on a final examination in
college algebra. Find the quartiles of the distribution</h2>

```python
import pandas as pd

data = {
    'Grade range': ['90-100', '80-89', '70-79', '60-69', '50-59', '40-49', '30-39'],
    'No. of students': [9, 32, 43, 21, 11, 3, 1]
}

df = pd.DataFrame(data)
df['Cumulative Frequency'] = df['No. of students'].cumsum()
n = df['No. of students'].sum()

q1_pos = (n + 1) / 4
q2_pos = 2 * (n + 1) / 4
q3_pos = 3 * (n + 1) / 4

def interpolate_quartile(position, df):
    for i, row in df.iterrows():
        if row['Cumulative Frequency'] >= position:
            prev_cum_freq = df['Cumulative Frequency'].iloc[i-1] if i > 0 else 0
            prev_grade_range = df['Grade range'].iloc[i-1] if i > 0 else ''
            
            total_students_in_range = row['No. of students']
            position_in_range = position - prev_cum_freq
            proportion = position_in_range / total_students_in_range
            
            lower_bound, upper_bound = map(int, row['Grade range'].split('-'))
            
            quartile_value = lower_bound + proportion * (upper_bound - lower_bound)
            return quartile_value
            
    return None

Quartile1_value = interpolate_quartile(q3_pos, df)
Quartile2_value = interpolate_quartile(q2_pos, df)
Quartile3_value = interpolate_quartile(q1_pos, df)

print(f"Q1 value is: {Quartile1_value:.2f}")
print(f"Q2 value is: {Quartile3_value:.2f}")
print(f"Q3 value is: {Quartile3_value:.2f}")
```

<h2>3. Prove that the mean and standard deviation of a set of standard scores are
equal to 0 and 1, respectively. 
Set = 6,2,8,7,5</h2>

```python
import numpy as np

numbers = np.array([6,2,8,7,5])
mean = sum(numbers)/5
std_ans = np.std(numbers, ddof=0)
z_score = (numbers-mean)/std_ans
z_score_mean = sum(z_score)/5
z_score_std = np.std(z_score, ddof=0)

print(f" Z-score Mean = {z_score_mean:.6f}")
print(f" Z-score Standard Deviation = {z_score_std}")
```
