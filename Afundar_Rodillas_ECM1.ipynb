```python
import random
sampleSize=["Marcos","Robredo","Marcos","Marcos","Robredo","Marcos"]
N=6

MarcosCount=0
RobredoCount=0
for i in range(len(sampleSize)):
    if sampleSize[i]=="Marcos":
            MarcosCount+=1
    else:
        RobredoCount+=1
        
print(MarcosCount/N)
print(RobredoCount/N)

n=4
checklist=[]
MarcosProp=[]
RobredoProp=[]

for j in range(15):
    sampleSize4=[]
    MarcosCount4=0
    RobredoCount4=0
    for i in range(4):
        sampleSize4=random.sample(sampleSize,4)
    for k in range(len(sampleSize4)):
        if sampleSize4[k]=="Marcos":
                MarcosCount4+=1
        else:
            RobredoCount4+=1
            
    checklist.append(sampleSize4)
    MarcosProp.append(MarcosCount4/n)
    RobredoProp.append(RobredoCount4/n)
    print("Sample:", j+1)
    print(sampleSize4)
    print("Marcos Proportion:",MarcosCount4/n)
    print("Robredo Proportion:",RobredoCount4/n)
    print("\n\n")
    
    
    
```

```python
import pandas as pd
from IPython.display import display


Electiondf = pd.DataFrame({
    'Sample': checklist,
    'Marcos': MarcosProp,
    'Robredo': RobredoProp
}) 
display(Electiondf)
```
